```python
import requests
import json
import pandas as pd
```

```python
keyset = {}
data = []
```

```python

for i in range(1,100) :
    url = f"https://comic.naver.com/api/challenge/list?genre=&order=VIEW&tab=TOTAL&page={i}"
    response = requests.get(url)
    json_data = response.json()
    # print(f"index : {i}")
    if 'list' in json_data :
        for item in json_data['list']:
            if 'greatestContest' in item and item['greatestContest'] == True  and item['titleId'] not in keyset:
                keyset[item['titleId']] = True
                data.append(item)
```

```python
for item in data :
    url = f"https://comic.naver.com/api/article/list/info?titleId={item['titleId']}"
    response = requests.get(url)
    json_data = response.json()
    item['favoriteCount'] = json_data['favoriteCount']
    item['synopsis'] = json_data['synopsis']
    item['tag'] = ''
    for tag in json_data['challengeTagList']:
        item['tag'] += tag + " "
    print(item['tag'])
```

```python
df = pd.DataFrame(data)
byview = df.sort_values(by=['favoriteCount'], axis=0, ascending=False)
```
